```python
import pandas as pd

df = pd.read_csv("/content/Games.csv")

df.head(10)
```

```python
import numpy as np

df['Score_by_devs'] = df['Score'] // 2
df.head()
```

```python
import matplotlib.pyplot as plt

plt.hist(bins=10, data=df, x='Score')
plt.title('Score histogram')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
```

```python
data = df['Review'].value_counts()
sum_of_others = 0
for val in data.iloc[4:]:
    sum_of_others+= val
sum_of_others

plt.figure(figsize=(10,6))


data_fileterd = data.iloc[:4]
data_fileterd['Others'] = sum_of_others
plt.subplot(121)
plt.pie(data_fileterd, labels=['Good', 'Great','Fair','Medicore','Others'], autopct='%1.1f%%', startangle=90)
plt.title('Review pie chart')
plt.show()


plt.figure(figsize=(20,6))
plt.subplot(122)
fig = plt.bar(data.iloc[4:].index, height=data.iloc[4:], color=['red', 'green','red','red','red','green','yellow'])

for rect in fig:
    height = rect.get_height()
    plt.annotate(f'{round((height/sum_of_others) * 100,2)}%', xy=(rect.get_x() + rect.get_width() / 2, height),
                 xytext=(0, 3), textcoords='offset points', ha='center', va='bottom')

plt.title('Others Category')
plt.show()
```
